## Solution:
1. **Binary Tree**:
A tree data structure in which each node has at most two children, which are referred to as the left child and the right child.

2. **Binary Search Tree (BST)**:
A binary tree where for each node, all elements in the left subtree are less than the node and all elements in the right subtree are greater than the node.

3. **Minimum Spanning Tree (MST)**:
A subset of the edges of a connected, edge-weighted undirected graph that connects all the vertices together, without any cycles and with the minimum possible total edge weight.

4. **Balanced Binary Tree**:
A binary tree in which the depth of the two subtrees of every node never differs by more than 1.

5. **Depth First Traversal (DFT) for Trees**:
A traversal method used to explore all the vertices of a graph or a tree in depth-ward motion. In trees, it can be implemented using recursion and for graphs, using a stack.

Use the Bellman-Ford algorithm to find the shortest path from node G to K in the weighted directed graph above. Show your work.

## Solution

````python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

G = nx.DiGraph()

nodes = ['G', 'H', 'I', 'J', 'K']
edges = [('G', 'H', 2), ('G', 'I', 4), ('H', 'I', -3), ('H', 'K', 3), ('I', 'J', 1), ('J', 'K', 2)]

G.add_nodes_from(nodes)
G.add_weighted_edges_from(edges)

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='skyblue', font_size=15, font_weight='bold')
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): G[u][v]['weight'] for u, v in G.edges()}, font_size=12)

plt.title("Directed Weighted Graph")
plt.show()

def bellman_ford(graph, source):
    # Step 1: Initialize distances and predecessors
    distance = {node: float('inf') for node in graph.nodes}
    predecessor = {node: None for node in graph.nodes}
    distance[source] = 0
    table = [distance.copy()]

    # Step 2: Relax edges repeatedly
    for _ in range(len(graph.nodes) - 1):
        for (u, v, weight) in graph.edges(data='weight'):
            if distance[u] + weight < distance[v]:
                distance[v] = distance[u] + weight
                predecessor[v] = u
        table.append(distance.copy())

    # Step 3: Check for negative-weight cycles
    for (u, v, weight) in graph.edges(data='weight'):
        if distance[u] + weight < distance[v]:
            raise ValueError("Graph contains a negative-weight cycle")

    return table, predecessor

table, _ = bellman_ford(G, 'G')

df = pd.DataFrame(table)
df.columns = [f"Dist to {col}" for col in df.columns]

df["Iteration"] = ["Initial"] + [f"Iteration {i}" for i in range(1, len(df))]

df.set_index("Iteration", inplace=True)

df
````

* **Dist to G, H, I, J, K:** These columns represent the computed shortest distances from the source node 'G' to the respective nodes after each iteration.
* **Iteration:** This column indicates the iteration of the algorithm.

From the table, we can observe the distances getting updated in the first iteration. Subsequent iterations do not cause any further updates as the shortest paths have already been established.

## New Problem:

### Problem Statement:
In a city, there are various interconnected roads represented by a directed graph $ G = (V,E) $ where the vertices represent intersections and the edges represent roads. Each road has a capacity which denotes the number of cars it can handle in an hour. There is a starting point $ s $ and a destination point $ t $. Due to a forecasted event, the city planners expect a certain number, say $ k $, of roads to be blocked. They wish to identify which $ k $ roads, when blocked, will cause the maximum disruption in traffic flow from $ s $ to $ t $. 

Given $ G $, $ s $, $ t $, and $ k $, design an algorithm to identify which $ k $ roads to block so as to minimize the traffic flow from $ s $ to $ t $.

### Input Format:
- A list of edges, where each edge is represented as a tuple $ (u, v, c) $ where $ u $ and $ v $ are the start and end points of the road respectively and $ c $ is the capacity of the road.
- The starting point $ s $.
- The destination point $ t $.
- An integer $ k $.

### Output Format:
- A list of $ k $ edges that should be blocked to minimize the flow from $ s $ to $ t $.

### Sample Input:

\begin{align*}
ext{Edges} & : [(1,2,10), (1,3,5), (2,3,15), (2,4,5), (3,4,10)] \
s & : 1 \
t & : 4 \
k & : 2 \
\end{align*}


### Sample Output:
$$
[(1,2,10), (2,3,15)]
$$
### Constraints:
1. $ 1 \leq |V| \leq 1000 $
2. $ 1 \leq |E| \leq 10^4 $
3. $ 1 \leq k \leq |E| $
4. $ 0 \leq c \leq 10^5 $



````python
import networkx as nx
from networkx.algorithms.flow import maximum_flow

def roads_to_block(edges, s, t, k):
    G = nx.DiGraph()
    for u, v, c in edges:
        G.add_edge(u, v, capacity=c)

    original_flow_value, _ = maximum_flow(G, s, t)

    delta_values = []
    for u, v, c in edges:
        G_temp = G.copy()
        G_temp.remove_edge(u, v)
        flow_value, _ = maximum_flow(G_temp, s, t)
        delta = original_flow_value - flow_value
        delta_values.append(((u, v, c), delta))

    delta_values.sort(key=lambda x: x[1], reverse=True)

    return [edge for edge, _ in delta_values[:k]]

edges = [(1,2,10), (1,3,5), (2,3,15), (2,4,5), (3,4,10)]
s = 1
t = 4
k = 2
G = nx.DiGraph()
for u, v, c in edges:
    G.add_edge(u, v, capacity=c)

pos = nx.spring_layout(G)
edge_labels = {(u, v): G[u][v]['capacity'] for u, v in G.edges()}
nx.draw(G, pos, with_labels=True, node_size=1000, node_color='lightblue', font_size=15, width=2, edge_color='gray')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=15)

plt.title("City Roads Graph")
plt.show()
roads_to_block(edges, s, t, k)
````

### Solution:
1. Calculate the maximum flow $ f $ from $ s$ to $ t $ in the original graph $ G $.
2. For each edge $ e $ in $ E $:
    - Remove $ e $ from $ G $ to form $ G' $.
    - Calculate the flow $ f' $ in $ G' $.
    - Compute the difference $ \delta = f - f' $.
    - Store $ \delta $ and associate it with edge $ e $.
3. Sort the edges based on $ \delta $ in descending order.
4. Return the top $ k $ edges.

This solution operates under the assumption that the removal of individual edges affects the flow value independently. 

### Proof of Correctness:
For every edge $ e $, the difference $ \delta $ represents how much the flow is reduced when $ e $ is removed. By sorting edges based on this value and selecting the top $ k $, we ensure that we're choosing the $ k $ roads that, when blocked, will cause the maximum disruption in traffic flow.

## New Question:
Imagine a scenario where you're organizing a series of concerts in a linear park. There are $N$ stages available in the park, each at a specific location. If you organize a concert on stage $i$, you will get a profit of $p_i$. However, due to sound interference, no two concerts can be held on adjacent stages.

Your goal is to organize concerts on a subset of the $N$ stages that maximizes the total profit, subject to the non-adjacency constraint.

You're organizing a series of concerts in a linear park with $N$ stages. For each stage $i$, if you organize a concert, you will earn a profit of $p_i$. Due to sound interference, no two concerts can be held on adjacent stages. Determine the maximum profit you can earn.

### Input Format:

- An integer $N (1\leq N \leq10^5)$ - the number of stages.
- A list of $N$ integers representing the profit for each stage.

### Output Format:

- An integer representing the maximum profit that can be earned.

### Sample Input:
5
6 7 8 9 10


### Sample Output:
24

### Constraints:

- The profit from each concert $p_i (1\leq p_i \leq 10^6)$.

### Solution:

1. **Define the Sub Problems**:
Let $OPT(i)$ be the maximum profit achievable from organizing concerts only on stages 1 . . . $i$.

2. **Recurrence Relation**:
$OPT(i) = \max(OPT(i - 1), p_i + OPT(i - 2))$

3. **Base Cases**:
$ OPT(0) = 0, OPT(1) = p_1 $

The logic behind this problem is similar to the ATM problem. You decide whether to hold a concert on stage $i$ or not. If you do, you can't hold a concert on stage $i-1$ due to the interference.


````python
def max_profit(n, profits):
    if n == 0:
        return 0
    if n == 1:
        return profits[0]
    
    opt = [0] * n
    opt[0] = profits[0]
    opt[1] = max(profits[0], profits[1])
    
    for i in range(2, n):
        opt[i] = max(opt[i-1], profits[i] + opt[i-2])
    
    return opt[n-1]

n = 5
profits = [6, 7, 8, 9, 10]
print(max_profit(n, profits))  # Outputs: 24
````

# New Question


## Problem Statement

You are planning a series of events at a venue. Each event has a start time, end time, and a profit associated with it. Due to venue restrictions, no two events can overlap. Your goal is to select a set of non-overlapping events that maximize the total profit. 

Given a list of events with their start times, end times, and profits, determine the maximum profit you can achieve and the list of events that should be selected.

### Input Format

- A list of `n` events where each event is represented by a tuple `(start_time, end_time, profit)`.
  - `0 <= n <= 1000`
  - `0 <= start_time < end_time <= 10000`
  - `1 <= profit <= 1000`

### Output Format

- An integer representing the maximum profit.
- A list of events that should be selected.

### Sample Input

```
[(1, 3, 5), (2, 5, 6), (4, 6, 5), (6, 9, 4), (5, 8, 3)]
```

### Sample Output

```
14
[(4, 6, 5), (6, 9, 4)]
```

### Constraints

- Events are given in no particular order.
- If there are multiple solutions with the same maximum profit, any one of them is valid.

---

## Solution:

````python
def weighted_interval_scheduling(events):
    sorted_events = sorted(events, key=lambda x: x[1])
    p = []
    for i in range(len(sorted_events)):
        j = i - 1
        while j >= 0:
            if sorted_events[j][1] <= sorted_events[i][0]:
                p.append(j)
                break
            j -= 1
        else:
            p.append(-1)  # No non-overlapping interval found

    # Dynamic Programming
    dp = [0] * (len(sorted_events) + 1)
    for i in range(1, len(sorted_events) + 1):
        if p[i-1] != -1:
            dp[i] = max(dp[i-1], sorted_events[i-1][2] + dp[p[i-1] + 1])
        else:
            dp[i] = max(dp[i-1], sorted_events[i-1][2])

    n = len(sorted_events)
    solution = []
    while n > 0:
        if (p[n-1] == -1 or dp[n] == sorted_events[n-1][2] + dp[p[n-1] + 1]):
            solution.append(sorted_events[n-1])
            if p[n-1] != -1:
                n = p[n-1] + 1
            else:
                n -= 1
        else:
            n -= 1

    solution.reverse()
    return dp[len(sorted_events)], [f"Event: {events.index(s)+1}" for s in solution]

events = [(1, 3, 5), (2, 5, 6), (4, 6, 5), (6, 9, 4), (5, 8, 3)]
weighted_interval_scheduling(events)
````

## Analysis:

The problem you've provided is the classic 0-1 Knapsack problem, which is a fundamental problem in combinatorial optimization. The goal is to determine the most valuable combination of items to include in a knapsack, given a weight constraint. The solution is found using dynamic programming, and it is built upon a table that keeps track of maximum values obtainable with $i$ items and weight $w$.

## New Question:

### Problem Statement:

Given a set of items, each with a weight and a value, determine the number of each item to include in a knapsack so that the total weight is less than or equal to a given limit and the total value is maximized. However, unlike the classic 0-1 Knapsack problem, you can take fractional amounts of items.

### Input Format:

- A list of $n$ tuples, where each tuple contains two values: the value $v_i$ and weight $w_i$ of the item.
- An integer $W$ representing the maximum capacity of the knapsack.

### Output Format:

- A list of $n$ floating-point numbers representing the fraction of each item to be included in the knapsack. A value of 1 means the entire item is taken, and a value of 0 means none of the item is taken.

### Sample Input:
\[
[(3, 2), (4, 3), (5, 4), (6, 5)], W=8
\]

### Sample Output:
\[
[1, 1, 0.5, 0]
\]

### Constraints:

- $1 \leq n \leq 1000$
- $1 \leq w_i, v_i \leq 1000$
- $1 \leq W \leq 10^4$

### Solution:

The fractional knapsack can be solved using a greedy approach. Sort the items by value-to-weight ratio. Then, take the items with the highest ratio first, until the knapsack is full. If the knapsack gets filled before considering all items, take the fractional part of the next item.

````python
def fractional_knapsack(items, W):
    # Sort items by value-to-weight ratio in descending order
    items_sorted = sorted(items, key=lambda x: x[0]/x[1], reverse=True)
    weight_accumulated = 0
    value_accumulated = 0
    fractions = []

    for item in items_sorted:
        if weight_accumulated + item[1] <= W:
            fractions.append(1)
            weight_accumulated += item[1]
            value_accumulated += item[0]
        else:
            fraction = (W - weight_accumulated) / item[1]
            fractions.append(fraction)
            weight_accumulated = W
            value_accumulated += item[0] * fraction
            break
    
    # Fill in the remaining items as 0
    while len(fractions) < len(items):
        fractions.append(0)
    
    # Return the fractions in the original order
    return [fractions[items_sorted.index(item)] for item in items]
def visualize_fractional_knapsack(items, W):
    fractions = fractional_knapsack(items, W)
    
    values, weights = zip(*items)
    
    fig, ax = plt.subplots(figsize=(10, 6))
    bars = ax.bar(range(len(items)), fractions, align='center', alpha=0.7, color='blue')
    
    ax.set_ylabel('Fraction Taken')
    ax.set_xlabel('Item Index')
    ax.set_title('Fraction of Each Item Taken in the Knapsack')
    ax.set_xticks(range(len(items)))
    ax.set_xticklabels([f'v={v}, w={w}' for v, w in items])
    ax.set_ylim(0, 1)
    
    # Annotate each bar with the fraction value
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
    
    plt.tight_layout()
    plt.show()

visualize_fractional_knapsack([(3, 2), (4, 3), (5, 4), (6, 5)], 8)
result = fractional_knapsack([(3, 2), (4, 3), (5, 4), (6, 5)], 8)
result
````
